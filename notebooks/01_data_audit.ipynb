```python
# Converting dates into correct format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Order Date'].head()  # [Transformed to: 2016-06-12; 2015-10-11]
# df['Ship Date'].head()
```

```python
# Key Performance Indicator(KPI)-1
# Calculating metric-1: 
# Lead Time(LT) = ship_date - order_date   The total no.of days btw ship and order dates

df['Lead_Time_Days'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Lead_Time_Days'].describe()
print(df['Lead_Time_Days'].head())
```

```python
# Saving the Cleaned file to processed folder
df.to_csv("D:/REVATURE/Project-1_SCIS/data/processed/superstore_cleaned.csv", index=False)
```
